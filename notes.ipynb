```python
df = pd.read_csv('data/games.csv')
df['date'] = pd.to_datetime(df['date'])
#df.insert(1, 'weekday', df['date'].dt.weekday)
df.insert(1, 'hour', df['date'].dt.hour)
df = df.astype({'hour': 'str'})
df.insert(1, 'elo_gap', df['player_elo'] - df['opponent_elo'])
column = 'elo_gap'
df[column] = df[column] / df[column].abs().max()
for col in ['mean_player_time_per_move', 'mean_opponent_time_per_move', 'moves', 'date', 'result', 'time_control', 'player_elo', 'opponent_elo']:
    df.drop(col, axis=1, inplace=True)
#df['time_class']= pd.factorize(df['time_class'])[0]
df = pd.get_dummies(df)
df['game_num'] = df.index
column = 'game_num'
df[column] = df[column] / df[column].abs().max()
df
```

```python
y_column = 'player_result'
train_count = 300

train_ds = df[:-300]
test_ds = df[-300:]

train_X = train_ds.drop([y_column], axis=1)
train_y = train_ds[y_column]
test_X = test_ds.drop([y_column], axis=1)
test_y = test_ds[y_column]
train_X
```

```python
X = df.drop('player_result', axis=1)
y = df['player_result']
import numpy as np
from sklearn.model_selection import train_test_split
X_train,X_test, y_train,  y_test = train_test_split(X,y, test_size=0.2, random_state=1234)
X_train = X_train.to_numpy()
y_train = y_train.to_numpy().astype(int)
X_test = X_test.to_numpy()
y_test = y_test.to_numpy().astype(int)
print(X_train.shape,y_train.shape, X_test.shape , y_test.shape)
y_test[:10]
X_test[0]
```

```python
import tensorflow.keras.models as models
import tensorflow.keras.layers as layers
import tensorflow.keras.utils as utils
import tensorflow as tf
import tensorflow.keras.optimizers as optimizers
```

```python
model = tf.keras.Sequential([
    
    tf.keras.layers.Input(shape=(X_train.shape[1])),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(16, activation='relu'),
    tf.keras.layers.Dense(16, activation='relu'),
    tf.keras.layers.Dense(16, activation='relu'),
    tf.keras.layers.Dense(16, activation='relu'),
    tf.keras.layers.Dense(1)
])
```

```python
model.compile(optimizer='adam',
              loss=tf.keras.losses.BinaryCrossentropy(),
              metrics=['accuracy'])
history = model.fit(X_train, y_train,
          epochs=40,
          verbose=1,
          validation_data=(X_test,y_test))
```

```python
test_loss, test_acc = model.evaluate(X_test,  y_test, verbose=2)

print('\nTest accuracy:', test_acc)
```
